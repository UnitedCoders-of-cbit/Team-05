```python
import pandas as pd
import numpy as np
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://startuptalky.com/indian-startups-funding-investors-data-2022/"
```

```python
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)
soup = BeautifulSoup(driver.page_source)
driver.close()
```

```python
tables=soup.find_all('table')
```

```python
tables
```

```python
rows=[]
for i in tables:
    rows.append(i.find_all('tr'))
```

```python
rows
```

```python
tds=[]
for i in rows:
    for j in i:
        tds.append(j.find_all('td'))
```

```python
tds
```

```python
tds[1]
```

```python
df=pd.DataFrame(columns=['Company Name','Founded','Location','Sector','What it does','Founders','Investor','Amount','Stage'])
```

```python
tds1= [ele for ele in tds if ele != []]
```

```python
for i in tds1:
    df.loc[len(df.index)] = i
```
